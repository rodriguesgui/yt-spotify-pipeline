```python
from sqlalchemy import create_engine, text   # text empacota a string SQL

engine = create_engine(
    "postgresql+psycopg2://postgres:1234@localhost:5432/postgres"
)

# opção A – usando text(...)
with engine.connect() as cnx:
    resultado = cnx.execute(text("SELECT 1")).scalar()
    print("Conectado! Resultado:", resultado)

# opção B – método específico do driver (dispensa text)
with engine.connect() as cnx:
    resultado = cnx.exec_driver_sql("SELECT 1").scalar()
    print("Conectado! Resultado:", resultado)


```

```python
df_new2.to_sql(
    'stg_spotify_youtube',
    engine,
    if_exists='replace',
    index=False,
    method='multi',
    chunksize=10_000
)
print('Tabela criada e dados carregados!')
```
